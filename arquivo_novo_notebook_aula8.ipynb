```python
def verificar_temperatura(temperatura):
    # Esta função verifica a temperatura de um sistema.
    # Se a temperatura for maior que 30°C, um alerta de superaquecimento é exibido.
    # Caso contrário, indica que a temperatura está normal.
    if temperatura > 30:
        # Exibe um alerta de superaquecimento usando concatenação de strings.
        print("ALERTA: Superaquecimento detectado! Temperatura atual: " + str(temperatura) + "°C")
    else:
        # Exibe uma mensagem de temperatura normal usando concatenação de strings.
        print("Temperatura normal. Temperatura atual: " + str(temperatura) + "°C")

# --- Exemplos de uso da função ---

# Simula uma temperatura acima do limite (32°C) para acionar o alerta.
sistema_temperatura = 32
verificar_temperatura(sistema_temperatura)

# Simula uma temperatura dentro da faixa normal (25°C).
sistema_temperatura_normal = 25
verificar_temperatura(sistema_temperatura_normal)
```

criar um programa que organize arquivos bagunçandos em pastas automaticamente, basendo-se apenas na extensão do arquivo(ex: .pdf, .jpg, .xls, .doc, .py, .ipynb).

use a biblioteca ose shutil.

comente a exe cucao durenre o codigo.

use 10arquivos de cada tipo.

```python
import os
import shutil
import random
import string

# --- Configurações Iniciais ---
diretorio_base = "./ArquivosBaguncados"

# Lista de extensões de arquivo a serem usadas para geração e organização
extensoes = [".pdf", ".jpg", ".xls", ".doc", ".py", ".ipynb"]

# Número de arquivos de cada tipo a serem gerados
num_arquivos_por_tipo = 10

# --- Função para gerar nomes de arquivos aleatórios ---
def gerar_nome_arquivo_aleatorio(length=8):
    """
    Gera um nome de arquivo aleatório com letras e números.
    """
    letras_numeros = string.ascii_lowercase + string.digits
    return ''.join(random.choice(letras_numeros) for i in range(length))

# --- Geração de Arquivos de Teste ---
print(f"Preparando o ambiente de teste em '{diretorio_base}'...")

# Cria o diretório base se ele não existir
if os.path.exists(diretorio_base):
    # Se o diretório já existe, remove seu conteúdo para garantir um teste limpo
    for item in os.listdir(diretorio_base):
        caminho_completo = os.path.join(diretorio_base, item)
        if os.path.isfile(caminho_completo):
            os.remove(caminho_completo)
        elif os.path.isdir(caminho_completo):
            shutil.rmtree(caminho_completo)
else:
    os.makedirs(diretorio_base)

print(f"Gerando {num_arquivos_por_tipo} arquivos de cada tipo em '{diretorio_base}'...")
for ext in extensoes:
    for i in range(num_arquivos_por_tipo):
        nome_arquivo = gerar_nome_arquivo_aleatorio() + ext
        caminho_arquivo = os.path.join(diretorio_base, nome_arquivo)
        # Cria um arquivo vazio (ou com conteúdo mínimo) para cada tipo
        with open(caminho_arquivo, 'w') as f:
            f.write(f"Conteúdo de teste para {nome_arquivo}")
        #print(f"  Criado: {nome_arquivo}")
print("Geração de arquivos de teste concluída.\n")

# --- Função principal de organização ---
def organizar_arquivos(diretorio_origem):
    """
    Organiza arquivos em pastas com base em suas extensões.

    Argumentos:
        diretorio_origem (str): O caminho do diretório contendo os arquivos a serem organizados.
    """
    # Verifica se o diretório de origem existe
    if not os.path.isdir(diretorio_origem):
        print(f"Erro: O diretório de origem '{diretorio_origem}' não existe.")
        return

    print(f"Iniciando a organização dos arquivos em '{diretorio_origem}'...\n")

    # Itera sobre todos os itens no diretório de origem
    for item in os.listdir(diretorio_origem):
        caminho_completo_item = os.path.join(diretorio_origem, item)

        # Verifica se o item é um arquivo (ignora diretórios)
        if os.path.isfile(caminho_completo_item):
            # Extrai o nome do arquivo e sua extensão
            nome_arquivo, extensao = os.path.splitext(item)

            # Converte a extensão para minúsculas e remove o ponto inicial, ex: '.pdf' -> 'pdf'
            extensao = extensao.lower().lstrip('.')

            # Se não houver extensão (arquivo sem ponto), move para uma pasta 'OUTROS'
            if not extensao:
                nome_pasta_destino = "OUTROS"
            else:
                # Cria o nome da pasta com base na extensão (ex: 'PDF', 'JPG')
                nome_pasta_destino = extensao.upper()

            # Constrói o caminho completo da pasta de destino
            caminho_pasta_destino = os.path.join(diretorio_origem, nome_pasta_destino)

            # Cria a pasta de destino se ela ainda não existir
            if not os.path.exists(caminho_pasta_destino):
                os.makedirs(caminho_pasta_destino)
                print(f"  Pasta criada: '{nome_pasta_destino}/'")

            # Constrói o caminho de destino do arquivo
            caminho_final_arquivo = os.path.join(caminho_pasta_destino, item)

            # Move o arquivo para a pasta de destino
            try:
                shutil.move(caminho_completo_item, caminho_final_arquivo)
                print(f"  Movido: '{item}' para '{nome_pasta_destino}/'")
            except shutil.Error as e:
                print(f"  Erro ao mover '{item}' (já existe?): {e}")
            except Exception as e:
                print(f"  Erro inesperado ao mover '{item}': {e}")
        else:
            # Ignora diretórios existentes, incluindo as pastas recém-criadas
            print(f"  Ignorando diretório: '{item}/'")

    print("\nOrganização de arquivos concluída com sucesso!")

# --- Execução do Programa ---
# Chama a função de organização passando o diretório base criado
organizar_arquivos(diretorio_base)
```
